# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=['Python','C','C++','Java']
y=[85,70,60,82]

c=['g','m','b','r'] # this list is used to change colors of the bars

plt.bar(x,y, width=0.4, color=c, # bar plot will take x axis and y axis as arguments
        align = 'edge',          # width changes the width of the bars (not the graph)
        edgecolor = 'r',         # change color of the bars
        linewidth = 10,          # align changes the labels of the bar to edge or center
        alpha = 0.4,             # edge color changes the edge colors of bars and linewidht changes the widht of the edges
        label = 'Popularity')    # alpha changes the transparency of graph and value is from 0 to 1 (float value)
                                 # label parameter will not be visible until we use legend() function

plt.xlabel('Language', fontsize = 10) # we can give label to x axis like this and we can change fontsize as well

plt.ylabel('Number') # we can give label to y axis like this

plt.title('Example') # we can give title like this

plt.legend()

plt.show()
```

### Defferentiating 2 graphs in 1

```python
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,50,40]

width=0.2

p = np.arange(len(x)) # numpy func which will create an array (not a list)
p1 = [j+width for j in p] #by adding width it will create one more bar right next to the existing one

plt.bar(p,y, width,label = 'Popularity')    #replaced x with p (p is an array created by using numpy)
plt.bar(p1,z, width,label = 'Popularity 1')    

plt.xlabel('Language', fontsize = 10) # we can give label to x axis like this and we can change fontsize as well

plt.ylabel('Number') # we can give label to y axis like this

plt.title('Example') # we can give title like this

plt.legend()

plt.xticks(p+width/2 , x) #xticks is used to get back the original names of x axis, p+width/2 is used to set position of the names

plt.show()
```

# Scatter Plot

```python
# Most parameters will be the same as bar plot like colors, title, xlabel, fontsize etc
# One different parameter we have here is 's' which resembles size of the dots
# Marker parameter will change the shape of the dots
# we also have a lot of color maps (google it)
# plt.colorbar() will provide a bar on the right side having different colors
# set_label() is a function which will give label to to any representation

x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,4,6]

sizes = [10,20,40,50,20,10,70]
plt.scatter(x,y, s = sizes, marker = '*')
plt.colorbar().set_label('Color bar') #gives a color bar

plt.title('Example', fontsize = 10)

plt.show()
```

# Histogram Plot

```python
import random
```

```python
x = np.random.randint(10,60,(50)) #only used for generating 50 numbers from 10 to 60 values
print(x)
```

```python
# Most parameters will be the same as bar plot like colors, title, xlabel, fontsize etc
# bins parameter is used 
'''A histogram displays numerical data by grouping data into "bins" of equal width. 
Each bin is plotted as a bar whose height corresponds to how many data points are in that bin. 
Bins are also sometimes called "intervals", "classes", or "buckets".'''
# Bottom parameter starts the y axis from that particular number
# orientation parameter changes the vertical to horizontal graph (orientation = 'horizontal')
# rwidth will increase the space between the graphs

no = [26,58,29,26,11,42,33,31,53,39,20,43,17,26,55,38,42,34,23,39,15,26,56,29,
      13,59,24,15,48,52,57,39,31,12,45,40,18,14,37,15,15,30,11,48,37,35,41,29,24,48]

plt.hist(no,bins = 10,bottom = 10, rwidth=0.8)

plt.show()
```

# Pie Plot

```python
# Most parameters will be the same as bar plot like colors, title, xlabel, fontsize etc
# autpct will show percentage of each section, f represents float value and '%0.1f%%' will be basic syntax
# radius will change the size of the pie chart
# startangle will change the starting point and helps to rotate the circle
# textprops accepts a dictionary and used to change the size of the label's font
# By default direction of the chart is counter clockwise, we can change it as well

x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.4,0.0,0.0,0.0] # used for showing a particular section separately

plt.pie(x, labels = y, explode = ex, autopct='%0.1f%%', radius = 0.9,startangle = 0, textprops={'fontsize':11}, counterclock=False)

plt.show()
```

```python
#play with the radius sizes to get appropriate chart

plt.pie(x, labels=y, radius = 1)
plt.pie([1],colors='w',radius = 0.5) # white pie chart inside the existing pie chart to get ring like structure

plt.show()
```

## Line Plot

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
```

### Using matplotlib to draw line graph

```python
plt.plot(var, var_1) 
plt.show() # seaborn does not have show attribute
```

### It is always better to use seaborn with DataFrames

```python
sns.lineplot(x=var,y=var_1) #using with list data
```

```python
x_1 = pd.DataFrame({'var':var,'var_1':var_1})

sns.lineplot(x='var',y='var_1',data=x_1)

# now we can see labels are coming by default

plt.show()
```

```python
y_1 = sns.load_dataset('penguins') #requires internet and will load data from https://github.com/mwaskom/seaborn-data
y_1
```

```python
y_1.head()
```

```python
# we don't need legend() function here to show hue
# style will change the style of the column inserted
# palette will give you access to multiple colours
# while using palette, pass 'A' keyword to get an error and it will provide the list of colours available
# markers will change the marker points
# dashes will remove the dotted line
# To remove legend, we use legend = False

sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y_1, hue='sex', 
             style='sex', 
             palette='Accent', 
             markers=['o','>'],
             dashes=False)
plt.show()
```

# Histogram Plot

```python
# histplot is the latest function used for histogram
# bins is used to organise data, look at the x axis numbers, compare both the graphs
# color will change the color of graph
# kde is kernel density estimate and will provide an estimation line (write kde=True)

sns.histplot(data=y_1, x='bill_length_mm')
plt.show()
```

```python
# working of bins function
# we can see the range is somwhere from 170 to 230, now let's use bins and check the difference btw both the graphs

sns.histplot(data=y_1, x='flipper_length_mm')
plt.show()
```

```python
# We can see the arrangement of the x axis is now according to the values given in the bins

sns.histplot(data=y_1, x='flipper_length_mm', bins=[170,180,190,200,210,220,230,240])
plt.show()
```
